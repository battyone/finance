```python
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

from pandas_datareader import data
from sklearn.preprocessing import MinMaxScaler
from matplotlib.pylab import rcParams
from keras.models import Sequential
from keras.layers import Dense, Dropout, LSTM
```

```python
# Define the instruments to download. We would like to see Apple, Microsoft and the S&P500 index.
tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'IBM', 'FB']

# We would like all available data from 01/01/2000 until 12/31/2016.
start_date = '2000-01-01'
end_date = '2018-12-25'

# User pandas_reader.data.DataReader to load the desired data. As simple as that.
apple_data = data.DataReader('AAPL', 'yahoo', start_date, end_date)
apple_data.head()
```

```python
# Creating dataframe
new_data = apple_data['Close'].to_frame()
train_size = math.floor(0.8*len(new_data))
```

```python
# Creating train and test sets
dataset = new_data.values

train = dataset[0:train_size,:]
valid = dataset[train_size:,:]
```

```python
# Converting dataset into x_train and y_train
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(dataset)

x_train, y_train = [], []
for i in range(60, len(train)):
    x_train.append(scaled_data[i-60:i,0])
    y_train.append(scaled_data[i,0])
x_train, y_train = np.array(x_train), np.array(y_train)

x_train = np.reshape(x_train, (x_train.shape[0],x_train.shape[1],1))
```

```python
# Create and fit the LSTM network
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(x_train.shape[1],1)))
model.add(LSTM(units=50, return_sequences=True, input_shape=(x_train.shape[1],1)))
model.add(LSTM(units=50))
model.add(Dense(1))

model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(x_train, y_train, epochs=1, batch_size=1, verbose=2)
```

```python
# Predicting 246 values, using past 60 from the train data
#inputs = new_data[len(new_data) - len(valid) - 60:].values
#inputs = inputs.reshape(-1,1)
#inputs  = scaler.transform(inputs)
inputs = new_data[len(new_data) - len(valid) - 60:]
len(new_data) - len(valid) - 60
```

```python

X_test = []
for i in range(60, inputs.shape[0]):
    X_test.append(inputs[i-60:i,0])
X_test = np.array(X_test)

X_test = np.reshape(X_test, (X_test.shape[0],X_test.shape[1],1))
closing_price = model.predict(X_test)
closing_price = scaler.inverse_transform(closing_price)
```

```python
% matplotlib inline

rcParams['figure.figsize'] = 20,10

train = new_data[:train_size]
valid = new_data[train_size:]
valid['Predictions'] = closing_price
plt.plot(new_data['Close'])
plt.plot(valid['Predictions'])
plt.title('Closing Price of Shares for Apple Inc.', fontsize=16, fontweight='bold')
plt.xlabel('Date', fontsize=12, fontweight='bold')
plt.ylabel('Closing Price', fontsize=12, fontweight='bold')
plt.legend(loc='upper left', prop={'size': 16})
valid = dataset[train_size:,:]
rms = np.sqrt(np.mean(np.power((valid - closing_price),2)))
print(f'Root Mean Squared Error: {rms}')
```

```python
# serialize model to JSON
model_json = model.to_json()
with open("apple_closing_price_model.json", "w") as json_file:
    json_file.write(model_json)
# serialize weights to HDF5
model.save_weights("apple_closing_price_model.h5")
print("Saved model to disk")
```
